```python
data_dir="C:\\*****\Class Codes"
import os
import pandas as pd
```

```python
os.chdir(data_dir)
dat=pd.read_csv("dm.csv")
```

```python
dat.isnull().sum()
```

```python
dat['History'].describe()
```

```python
dat['History'].unique()
```

```python
dat['History'].fillna('Missing',inplace=True)
```

```python
X=dat.drop(["AmountSpent","Cust_Id"],axis=1)
```

```python
#Create Dummies
X=pd.get_dummies(X,dtype=int)
X
```

```python
y=dat['AmountSpent']
```

```python
import sklearn.model_selection as model_selection
X_train,X_test,y_train,y_test=model_selection.train_test_split(X,y,test_size=0.2,random_state=200)
```

```python
#Bagging Regressor
from sklearn.ensemble import BaggingRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python
clf=BaggingRegressor(oob_score=True,n_jobs=-1,n_estimators=20,random_state=400,
                      base_estimator=DecisionTreeRegressor())
```

```python
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test) #Average number of correct predictions
```

```python
clf.oob_score_
```

```python
#Parameter tuning
for w in range(100,200,20):
    clf=BaggingRegressor(oob_score=True,n_jobs=-1,n_estimators=w,random_state=400,
                          base_estimator=DecisionTreeRegressor())
    clf.fit(X_train,y_train)
    oob=clf.oob_score_
    print('For n_estimators = '+str(w))
    print('OOB score is '+str(oob))
    print('************************')
```

```python
#Finalizing on a tree model with 160 trees
clf=BaggingRegressor(oob_score=True,n_jobs=-1,n_estimators=160,random_state=400,
                      base_estimator=DecisionTreeRegressor())
clf.fit(X_train,y_train)
```

```python
clf.score(X_test,y_test) #Average Number of correct predictions
```

```python
# Feature Importance
clf.estimators_
```

```python
len(clf.estimators_)
```

```python
# We can extract feature importance from each tree then take a mean for all trees
import numpy as np
imp=[]
for i in clf.estimators_:
    imp.append(i.feature_importances_)
imp=np.mean(imp,axis=0)
```

```python
feature_importance=pd.Series(imp,index=X.columns.tolist()).sort_values(ascending=False)
feature_importance
```
